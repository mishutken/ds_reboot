```python
#Задача 1
import pandas as pd
import requests
import io
import re
l_url ="https://data.gov.ru/opendata/7708660670-rodnik-neskuchniy-sad/data-20160608T1215-structure-20160608T1215.csv"
l_s = requests.get(l_url).content
l_df = pd.read_csv(io.StringIO(l_s.decode('UTF8')))
display(l_df)

l_df.index.name = "Показатель"
l_expr_1 = "в пределах "
l_expr_2 = "не более "


for i, l_str in l_df.iterrows():
    #проверка столбца анализ данных на предмет отсутствия float
    l_rez_s = re.findall(r'\d+\,\d+|\d+\.\d+|\d+',l_df.at[i, "Результат анализа"])
    try:
        l_rez_f = float(l_rez_s[0])
    except:
        l_rez_f = float(input(l_df.at[i, "Показатель"] + " не найден, введите ЧИСЛО = ")) #б/цвета = 0   
    #считываем норматив
    l_str_normat = l_df.at[i, "Норматив"]
    #Забираем предельные показатели, подгатавливаем к конвертации
    l_str_normat_2 = re.findall(r'\d+\,\d+|\d+\.\d+|\d+',l_str_normat)
    l_list_normat = []
    for j in l_str_normat_2:
        l_list_normat.append(float(re.sub(',', '.', j)))

    if l_str_normat.find(l_expr_1) >= 0:
        if l_rez_f >= float(l_list_normat[0]) and l_rez_f <= float(l_list_normat[1]):
            l_df.at[i, 'Вывод'] = 'Норма'
        else:
            l_df.at[i, 'Вывод'] = 'Отклонение'
    elif l_str_normat.find(l_expr_2) >= 0:
        if l_rez_f <= float(l_list_normat[0]):
            l_df.at[i, 'Вывод'] = 'Норма'
        else:
            l_df.at[i, 'Вывод'] = 'Отклонение'

display(l_df)


```

```python
#Задача 2
#В ящике 5 апельсинов и 4 яблока. Наудачу выбираются 3 фрукта. Какова вероятность, что все три фрукта – апельсины? 0.119209
#В интернете полученный аналитически ответ 0.119. Подтверждается ли он эксперементально? Да

import numpy as np
import random
l_iter = 1000000
l_success = 0

for q in range(l_iter):
    #Наполняем корзину фруктами
    l_busket = []
    l_orange = 5    # 1
    l_apple = 4     # 0

    for i in range(l_orange):
        l_busket.append(1)
    for i in range(l_apple):
        l_busket.append(0)

    l_count = 3     # кол-во попыток
    l_pozitive = 0  # кол-во полученных апельсинов
    for i in range(l_count):
        l_index = random.randint(0, len(l_busket)-1)
        if l_busket[l_index] == 1:
            l_pozitive += 1
        l_busket.pop(l_index)
        
    if l_pozitive == 3:
        l_success += 1

print("попыток проведено = ", l_iter, "\nвероятность =", l_success / l_iter)
```

```python
#Задача 3
#Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. 
#Какова вероятность, что он проверит ровно две детали? 0.233595
#В интернете полученный аналитически ответ 7/30 или 0.23333. 
#Подтверждается ли он эксперементально? Да
import numpy as np
l_success_shuf = 0
l_success_rand = 0
l_iter = 1000000
l_item_all = 10
l_item_non = 3

l_items = []
for i in range(l_item_non):
    l_items.append(0)
for i in range(l_item_all - l_item_non):
    l_items.append(1)
print(l_items)
#метод перемешивания выдает вероятность = 0.233595
for i in range(l_iter):
    np.random.shuffle(l_items)
    if l_items[0] == 0 and l_items[1] == 1:
        l_success_shuf += 1
print("shuffle проведено = ", l_iter, "\nвероятность =", l_success_shuf / l_iter)

#метод randint выдает вероятность = 0.233153
for q in range(l_iter):

    l_items = []
    for i in range(l_item_non):
        l_items.append(0)
    for i in range(l_item_all - l_item_non):
        l_items.append(1)

    l_index = random.randint(0, len(l_items)-1)
    if l_items[l_index] == 0:
        l_items.pop(l_index)
        l_index = random.randint(0, len(l_items)-1)
        if l_items[l_index] == 1:
            l_success_rand += 1
print("randint проведено = ", l_iter, "\nвероятность =", l_success_rand / l_iter)



```
